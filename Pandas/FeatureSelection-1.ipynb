# Feature Selection- Dropping the constant features

```python
import pandas as pd
import numpy as np
```

```python
data = pd.DataFrame({'A' : [1,2,4,1,2,4],
                     'B' : [4,5,6,7,8,9],
                     'C' : [0,0,0,0,0,0],
                     'D' : [1,1,1,1,1,1]})
data.head()
```

# Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
vartreshold = VarianceThreshold(threshold=0.0)
vartreshold.fit(data)
```

```python
data.columns
```

```python
data.columns[vartreshold.get_support()]
```

```python
const_columns = [x for x in data.columns if x not in data.columns[vartreshold.get_support()]]
const_columns
```

```python
data.drop(const_columns, axis=1, inplace=True)
data
```
