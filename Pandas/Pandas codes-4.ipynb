```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x:x.str.contains('Python').sum())
country_uses_python
```

### Now that we hav 2 Series so use .concat method to concat them ....and if you want to append one below another then use the .append() method

```python
python_df = pd.concat([country_respondents, country_uses_python], axis='columns')
python_df
```

```python
python_df = python_df.rename(columns={'Country' : 'NumRespondents', 'LanguageWorkedWith' : 'NumKnowsPython'})
python_df
```

```python
python_df['PercentageKnowsPython'] = (python_df['NumKnowsPython']/python_df['NumRespondents'])*100
python_df
```

```python
python_df.sort_values(by='PercentageKnowsPython', ascending=False, inplace=True)
```
